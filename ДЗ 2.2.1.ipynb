```python
import numpy as np 
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df.shape
```

```python
df.head(5)
```

```python
sns.regplot(data=df, x='capital', y='output', ci=None, line_kws={"color": "r"})
```

```python
fitted_line = smf.ols(formula='output~capital', data=df).fit()
fitted_line.params.round(2)
```

## Ответ: Intercept = 6.19; capital = 0.74 (параметры оптимальной прямой output на capital)

```python
fitted_line = smf.ols(formula='output~-1+capital', data=df).fit()
fitted_line.params.round(2)
```

## Ответ: capital = 0.75 (параметры оптимальной прямой output на capital без константы)

```python
sns.regplot(x=np.log(df['capital']), y=np.log(df['output']), ci=None, line_kws={"color": "r"})
```

```python
fitted_line = smf.ols(formula='np.log(output)~np.log(capital)', data=df).fit()
fitted_line.params.round(2)
```

## Ответ: Intercept = 1.19; np.log(capital) = 0.56 (найти параметры оптимальной прямой log(output) на log(capital))

```python
fitted_line = smf.ols(formula='np.log(output)~-1+np.log(capital)', data=df).fit()
fitted_line.params.round(2)
```

## Ответ: Intercept = 1.19; np.log(capital) = 0.56 (найти параметры оптимальной прямой log(output) на log(capital) без константы)
