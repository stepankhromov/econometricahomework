```python
import numpy as np 
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df.shape
```

```python
sns.regplot(x=np.log(df['labour']), y=np.log(df['output']), ci=None, order=2, line_kws={"color": "r"})
```

```python
fitted_parabola = smf.ols(formula='np.log(output)~np.log(labour)+I(np.log(labour)**2)', data=df).fit()
fitted_parabola.params.round(2)
```

## Ответ: Intercept = -1.28; np.log(labour) = 0.37; I(np.log(labour) ** 2)  = 0.06 (параметры оптимальной параболы log(output) vs log(labour), log2(labour))
