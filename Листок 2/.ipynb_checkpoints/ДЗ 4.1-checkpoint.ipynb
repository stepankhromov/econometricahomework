```python
import numpy as np 
import pandas as pd 
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Electricity.csv')
df
```

# Спецификация
$$
log(cost) = \beta_0 + \beta_1*log(q)
$$

```python
cost = smf.ols(formula='np.log(cost)~np.log(q)', data=df).fit()
cost.params.round (3)
```

$$
log(cost) = -3.841 + 0.836*log(q)
$$

При увеличении общих издержек за год на 1%, общий выпуск электроэнергии увеличивается на 83,6%
