```python
import pandas as pd
import numpy as np
import io
import statsmodels.formula.api as smf
```

```python
pd.set_option('display.max_columns', None)
# показывает все столбцы в таблице
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
# Можно пользоваться такой записью
df[df['smsa']>=1]
```

```python
# Есть такой вариант фильтрации. Его преимущество в том, что можно проделывать вычисление и накладывать фильтр. 
#Тот же результаты можно получить, воспользовавашись методом .apply()(об этом методе мы поговорим на следуюещм семинаре) но скорость метода .loc будет выше.
df.loc[df.smsa>=1].head(10)
```

```python
df = df[df.smsa >= 1]
df
# удалил все строки, которые НЕ smsa=0
```

```python
sleep_eq1 = smf.ols(formula='sleep~totwrk+age+male+south', data=df).fit()
sleep_eq1.params
```

```python
sleep_eq1.ess.round(3)
# ESS
```

```python
sleep_eq1.centered_tss.round(3)
# TSS
```

```python
RSS=sleep_eq1.ssr.round(3)
RSS
# RSS
```

```python
sleep_eq1.rsquared.round(7)
# R2
```

```python
sleep_eq1.rsquared_adj.round(7)
# R2adj
```

```python
n=282
k=4
s=np.sqrt(RSS/(n-k-1))
round(s,4)
```
