# 2. Формулировка экономической модели

### Проанализируем влияние _возраста, пола, индекса массы тела, количества детей в семье, а также курения_ на стоимость медицинского страхования:

$$ charges = \beta_0 + \beta_1*age + \beta_2*sex + \beta_3*bmi + \beta_4*children + \beta_5*smoker$$

```python
model = smf.ols('charges ~ age + C(sex) + bmi + children + C(smoker)',data = df).fit()
model.summary()
```

# 3. Интерпретация коэффициентов

#### Пол:
$\bullet$ $\textit{Мужчины}$ в среднем тратят на страховку на 128.6399 долларов $\textit{меньше, чем женщины}$;

$\bullet$ Значение коэффициента $\textit{статистически не значимо}$ (p_value = 0.700), то есть, $\textit{пол не оказывает статистически значимого влияния на стоимость страховки}$ при учете других переменных в модели.

#### Курение:
$\bullet$ $\textit{Курящие люди}$ имеют в среднем стоимость страховки на 23820 долларов $\textit{выше, чем некурящие люди;}$

$\bullet$ Значение коэффициента $\textit{статистически значимо}$ (p_value = 0.000), что подтверждает $\textit{значительное влияние курения на стоимость страховки}$ при учете других переменных в модели;

$\bullet$ Это может быть объяснено тем, что $\textit{курение существенно влияет на здоровье}$ и $\textit{повышает риски развития различных заболеваний,}$ что отражается на стоимости медицинской страховки.

#### Возраст:
$\bullet$ Каждый год возраста $\textit{увеличивает стоимость страховки}$ примерно на 257.735 долларов;

$\bullet$ Значение коэффициента $\textit{статистически значимо}$ (p_value = 0.000), что подтверждает $\textit{значительное влияние возраста}$ на стоимость страховки при учете других переменных в модели;

$\bullet$ Это может быть объяснено тем, что с $\textit{возрастом увеличивается риск возникновения различных заболеваний}$ и потребность в медицинском обслуживании, что отражается в стоимости медицинской страховки.

#### Индекс массы тела:
$\bullet$ При увеличении индекса массы тела на одну единицу, $\textit{стоимость страховки увеличивается}$ примерно на 322.3642 долларов;

$\bullet$ Значение коэффициента $\textit{статистически значимо}$ (P_value = 0.000), что указывает на наличие значительного влияния индекса массы тела на стоимость страховки при учете других переменных в модели;

$\bullet$ Это означает, что $\textit{люди с более высоким индексом массы тела склонны платить больше}$ за медицинскую страховку.

#### Количество детей:
$\bullet$ При увеличении количества детей на одного человека, $\textit{стоимость страховки увеличивается}$ примерно на 474.4111 долларов;

$\bullet$ Значение коэффициента $\textit{статистически значимо}$ (p_value = 0.000), что подтверждает $\textit{значительное влияние количества детей}$ на стоимость страховки при учете других переменных в модели;

$\bullet$ Это может быть объяснено тем, что $\textit{дети требуют обслуживания и заботы, включая регулярные посещения врачей,}$ что отражается в стоимости медицинской страховки.

# 6. Анализ модели на гетероскедастичность

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/3insurance.csv')
df
```

```python
df['res2']=model.resid**2
df['res']=model.resid
df
```

```python
model = smf.ols('charges ~ age + C(sex) + bmi + children + C(smoker)',data = df).fit()
```

```python
submodel=model = smf.ols('res2 ~ age + C(sex) + bmi + children + C(smoker)',data = df).fit()
```

```python
St = np.round(len(df)*submodel.rsquared,2)
St
```
