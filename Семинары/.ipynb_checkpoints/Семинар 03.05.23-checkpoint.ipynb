```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ryupepa/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
model = smf.ols(data=df, formula='sleep~totwrk+age+I(age**2)+male+smsa+south').fit()
```

```python
# квадраты остатков добавим в датафрейм, воспользовавшись методом '.resid', возвращающий значения остатков
df['res2']=model.resid**2
df['res']=model.resid
```

```python
sub_model= smf.ols(data=df, formula='res2~totwrk+age+I(age**2)+male+smsa+south').fit()
```

```python
df
```

```python
St = np.round(len(df)*sub_model.rsquared,2) #статистика, которую необходимо сравнивать с хи-квадрат
St #наблюдаемое
```

```python
Hi2 = stats.chi2.ppf(1-0.05,6) #(доверительный интервал, количество регрессоров, которые проверяем)
Hi2 #критическое
```

# Задача 1.1

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/icecream.csv')
df
```

```python
mdl=smf.ols(data=df, formula = 'cons~income+price+temp').fit()
```

```python
mdl.params.round(2)
```

```python
# вычислим статистику DW
```

```python
# вычислим знаменатель
resSq = sum(mdl.resid**2)
```

```python
# Вычислим числительно
sum_resid = 0 #
for i in range(len(mdl.resid)-1):
    sum_resid+=(mdl.resid[i+1]-mdl.resid[i])**2 # +=исходное сум_резид увеличиваем каждый раз эту разность
```

```python
sum_resid/resSq #статистика DW в тесте, автокорелляция положительная, гипотеза отвергается
```
