```python
import pandas as pd
import numpy as np
import seaborn as sns
import io
import statsmodels.formula.api as smf
import scipy.stats as stats
```

# Задача 4.1

```python
# Определим датафрейм
sleep_df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
sleep_df
```

```python
sleep_df['sleep60'] = sleep_df['sleep']/60 # создадим столбец часов времени
sleep_df
```

```python
sleep_male = sleep_df[sleep_df['male']==1] # Датафрейм для мужчин
sleep_female = sleep_df[sleep_df['male']==0] # Датафрейм дла женщин
```

```python
slp = smf.ols(data= sleep_df, formula='sleep60~totwrk+age+I(age**2)+south +smsa +marr').fit()
slp_ml = smf.ols(data= sleep_male, formula='sleep60~totwrk+age+I(age**2)+south +smsa +marr').fit()
slp_fml = smf.ols(data= sleep_female, formula='sleep60~totwrk+age+I(age**2)+south +smsa +marr').fit()
```

```python
print(f'RSS:{slp.ssr.round(2)}, RSS Male:{slp_ml.ssr.round(2)},RSS Female:{slp_fml.ssr.round(2)} ') #все, что находится в кавычках - строка, в фигурной скобке - вычисленное значение
```

```python
#вычислили остаточные суммы квадратов
```

```python
F_nabl =( slp.ssr - (slp_ml.ssr + slp_fml.ssr) )/((slp_ml.ssr + slp_fml.ssr)) * (len( sleep_female ) + len( sleep_male ) - 14) / 7
round(F_nabl,3)
#считаем F_nabl по формуле
```

```python
alpha = 0.01
F_cr=stats.f.ppf(1-alpha,7,706-7) 
round(F_cr,3)
```

Вывод: $F_{obs}<F_{crit}$, значит данные согласуются с нулевой гипотезой $H_0$. Структурные различия незначимы.

```python
sns.lmplot(data=sleep_df, y='sleep', x='totwrk',ci=None, hue='male') #ci=None - убрать доверительный интервал с графика
```
